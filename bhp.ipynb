```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('bengaluru_house_prices.csv')
df.head()
```

```python
df.shape
```

```python
df.groupby('area_type')['area_type'].agg('count')
```

```python
df.drop(['area_type', 'society', 'availability'], axis='columns', inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df[df['location'].isnull()]
```

```python
df.drop(df[df['location'].isnull()].index, inplace=True)
```

```python
df['balcony'].fillna(0, inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df['size'].unique()
```

```python
df['bhk'] = df['size'].str.split(' ').str[0]
df.head()
```

```python
df2 = df.dropna()
df2.head()
```

```python
df2.drop(['size'], axis='columns', inplace=True)
df2.head()
```

```python
df2['total_sqft'].unique()
```

```python
def total_sqft_to_float(x):
    try:
        tokens = x.split('-')
        if len(tokens) == 2:
            return (float(tokens[0]) + float(tokens[1])) / 2
        return float(x)
    except:
        return None
```

```python
df3 = df2.copy()
df3['total_sqft'] = df3['total_sqft'].apply(total_sqft_to_float)
df3.head()
```

```python
df4 = df3.copy()
df4['price_per_sqft'] = (df4['price'] * 100000 / df4['total_sqft']).round(2)
df4.head()
```

```python
df4.location.unique()
```

```python
df4.location = df4.location.apply(lambda x: x.strip())
df5 = df4.copy()
location_stats = df4.groupby('location')['location'].agg('count').sort_values(ascending=False)
df5.location = df5.location.apply(lambda x: 'other' if location_stats[x] <= 10 else x)
df5.head()
```

```python
df5['bhk'] = df5['bhk'].astype(int)
df5[df5.total_sqft / df5.bhk < 300].head()
```

```python
df6 = df5[~(df5.total_sqft / df5.bhk < 300)]
df6.shape
```

```python
df6.price_per_sqft.describe()
```

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft > (m - st)) & (subdf.price_per_sqft <= (m + st))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    return df_out
df7 = remove_pps_outliers(df6)
df7.shape
```

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk - 1)
            if stats and stats['count'] > 5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft < (stats['mean'])].index.values)
    return df.drop(exclude_indices, axis='index')
df8 = remove_bhk_outliers(df7)
df8.shape
```

```python
df9 = df8[df8['bath'] < df8['bhk'] + 2]
df9.shape
```

```python
df9.drop(['price_per_sqft'], axis='columns', inplace=True)
df9.head()
```

```python
dummies = pd.get_dummies(df9.location, dtype=int)
```

```python
df10 = pd.concat([df9.drop('location', axis='columns'), dummies.drop('other', axis='columns')], axis='columns')
df10.head()
```

```python
df10.shape
```

```python
X = df10.drop(['price'], axis='columns')
y = df10.price
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
lr_model.score(X_test, y_test)
```

```python
def predict_price(location, total_sqft, bath, bhk, balcony):
    loc_index = np.where(X.columns == location)[0][0]
    x = np.zeros(len(X.columns))
    x[0] = total_sqft
    x[1] = bath
    x[2] = balcony
    x[3] = bhk
    if loc_index >= 0:
        x[loc_index] = 1
    return lr_model.predict([x])[0]
```

```python
prediction = predict_price('1st Phase JP Nagar', 1000, 2, 2, 1)
print(f"Prediction: {prediction:.2f}")
```
