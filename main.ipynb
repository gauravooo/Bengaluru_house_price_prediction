```python
df['bhk'] = df['size'].str.split(' ').str[0]
df.head()
```

```python
df2 = df.dropna()
df2.head()
```

```python
df2.drop(['size'], axis='columns', inplace=True)
df2.head()
```

```python
df2['total_sqft'].unique()
```

```python
def total_sqft_to_float(x):
    try:
        tokens = x.split('-')
        if len(tokens) == 2:
            return (float(tokens[0]) + float(tokens[1])) / 2
        return float(x)
    except:
        return None
```

```python
df3 = df2.copy()
df3['total_sqft'] = df3['total_sqft'].apply(total_sqft_to_float)
df3.head()
```

```python
df4 = df3.copy()
df4['price_per_sqft'] = (df4['price'] * 100000 / df4['total_sqft']).round(2)
df4.head()
```

```python
df4.location.unique()
```

```python
df4.location = df4.location.apply(lambda x: x.strip())
df5 = df4.copy()
location_stats = df4.groupby('location')['location'].agg('count').sort_values(ascending=False)
df5.location = df5.location.apply(lambda x: 'other' if location_stats[x] <= 10 else x)
df5.head()
```
